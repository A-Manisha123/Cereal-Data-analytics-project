```python
# Scatterplot of calories vs. protein
plt.figure(figsize=(8, 6))
plt.scatter(x = 'calories', y = 'protein', data = 'df')
plt.title('calories vs. protein')
plt.xlabel('calories per serving')
plt.ylabel('protein per serving (g)')
plt.show()
```

```python
# Histogram of sugar content
plt.figure(figsize=(8, 6))
sns.histplot(df['sugars'], kde=True)
plt.title('Sugar Content Distribution')
plt.xlabel('Sugar per serving (g)')
plt.ylabel('Frequency')
plt.show()
```
